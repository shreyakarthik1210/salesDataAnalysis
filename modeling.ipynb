```python
from datetime import datetime

def toDateTime (data: pd.DataFrame = None, column: str = None):
  copy = data.copy()
  print(copy[column])
  copy[column] = pd.to_datetime(copy[column], format='%Y-%m-%d %H:%M:%S')
  return copy

sales_df = toDateTime(sales_df, 'timestamp')
sales_df.info()

stock_df = toDateTime(stock_df, 'timestamp')
stock_df.info()

temp_df = toDateTime(temp_df, 'timestamp')
temp_df.info()
```

```python
from datetime import datetime

def timeStampToHour(data: pd.DataFrame = None, col: str = None):
  copy = data.copy()
  new_ts = copy[col].tolist()
  new_ts = [i.strftime('%Y-%m-%d %H:00:00') for i in new_ts]
  new_ts = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in new_ts]
  copy[col] = new_ts
  return copy
```

```python
sales_df = timeStampToHour(sales_df, 'timestamp')
stock_df = timeStampToHour(stock_df, 'timestamp')
temp_df = timeStampToHour(temp_df, 'timestamp')
```

```python

df1 = stock_df.groupby(["timestamp", "product_id"]).agg({'estimated_stock_pct': 'mean'}).reset_index()
df2 = sales_df.groupby(["timestamp", "product_id"]).agg({'quantity': 'sum'}).reset_index()
df3 = temp_df.groupby("timestamp").agg({'temperature': 'mean'}).reset_index()
df3.head()
```

```python
merged = df1.merge(df2, on=["timestamp", "product_id"], how="left")
merged = merged.merge(df3, on="timestamp", how="left")
merged.head()
#merged.describe()
```

```python
merged['quantity'] = merged['quantity'].fillna(0)
#merged["temperature"] = merged["temperature"].fillna(-0.207)
products = sales_df[["product_id", "category"]].drop_duplicates()
merged = pd.merge(merged, products, on="product_id", how="left")
merged.head()
```

```python
price = sales_df[['product_id', 'unit_price']].drop_duplicates()
merged = pd.merge(merged, price, on="product_id", how="left")
merged.head()
merged.info()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  from matplotlib import pyplot as plt
  import seaborn as sns
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['timestamp']
  ys = series['estimated_stock_pct']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = _df_48.sort_values('timestamp', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('timestamp')
_ = plt.ylabel('estimated_stock_pct')
```
